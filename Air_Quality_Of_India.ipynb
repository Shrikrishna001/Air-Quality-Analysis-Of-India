```python
#checked unique values for state column
np.unique(df['state'])
```

```python
#Drop unnecessary columns
df.drop(['stn_code','location_monitoring_station','date'], axis='columns')
```

```python
#Simplified the column name by renaming it for better understanding
final_df = df
final_df.rename(columns={'sampling_date':'sampling_collect_date', }, inplace = True)
df.set_index('sampling_collect_date',inplace=True)
```

```python
#Total no of recored in each state of the state column
df['state'].value_counts()
```

```python
#in this we have plotted frequencies of various state using histogram
plt.figure(figsize=(50, 10))
plt.xticks(rotation=90)
df.state.hist()
plt.xlabel('state')
plt.ylabel('Frequencies')
plt.plot()
```

```python
#Types of areas & their values counts
df['type'].value_counts()
```

```python
#in this we have plotted frequencies of various areas types using histogram
plt.figure(figsize=(20, 6))
plt.xticks(rotation=90)
df.type.hist()
plt.xlabel('type')
plt.ylabel('Frequencies')
plt.plot()
```

```python
#This column describe which agency State Pollution Control Board has collected the most sample
df['agency'].value_counts()
```

```python
#In this we have plotted frequencies of various agency using histogram
plt.figure(figsize=(50, 10))
plt.xticks(rotation=90)
df.agency.hist()
plt.xlabel('agency')
plt.ylabel('Frequencies')
plt.plot()
```

```python
#In this visulization we've shown which state had the highest 'so2'
plt.figure(figsize=(30, 10))
plt.xticks(rotation=90)
sns.barplot(x='state',y='so2',data=df)
```

```python
#In this visulization we've shown which state had the highest 'so2' in ascending order
plt.rcParams['figure.figsize']=(30,10)
df[['so2','state']].groupby(["state"]).mean().sort_values(by='so2').plot.bar(color='purple')
plt.show()
```

```python
#In this visulization we've shown which state had the highest 'no2' 
plt.figure(figsize=(30, 10))
plt.xticks(rotation=90)
sns.barplot(x='state',y='no2',data=df);
```

```python
#In this visulization we've shown which state had the highest 'no2' in acending order
plt.rcParams['figure.figsize']=(30,10)
df[['no2','state']].groupby(["state"]).mean().sort_values(by='no2').plot.bar(color='purple')
plt.show()
```

```python
#In this visulization we've shown which state had the highest 'rspm'
plt.figure(figsize=(30, 10))
plt.xticks(rotation=90)
sns.barplot(x='state',y='rspm',data=df)
```

```python
#In this visulization we've shown which state had the highest 'rspm' in acending order
plt.rcParams['figure.figsize']=(30,10)
df[['rspm','state']].groupby(["state"]).mean().sort_values(by='rspm').plot.bar(color='purple')
plt.show()
```

```python
#In this visulization we've shown which state had the highest 'spm'
plt.figure(figsize=(30, 10))
plt.xticks(rotation=90)
sns.barplot(x='state',y='spm',data=df)
```
